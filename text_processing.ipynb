# Text Processing

*To tokenize the tweets,remove stopwords,remove whitespaces,contractions and lemmatize the words obtained after tokenizing*

```python
import pandas as pd
import string
import re
import contractions
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords,wordnet
from nltk.stem import WordNetLemmatizer
```

```python
df = pd.read_csv('data3.csv',encoding = 'ISO-8859-1')
```

```python
pd.set_option('display.max_colwidth',-1)
df.head(22)
```

```python
df = df.drop('cleaned_data',axis=1)
```

```python
#converting all words to lower case
df['final_data'] = df['final_data'].apply(lambda x: str(x).lower())
```

```python
df.head(200)
```

```python
#removing whitespaces
def rem_whitespaces(text):
    text = re.sub('\s+',' ',str(text),flags=re.MULTILINE)
    return text 
df['nospace'] = df['final_data'].apply(rem_whitespaces)
```

```python
#removing word contractions
df['no_contract'] = df['nospace'].apply(lambda x : [contractions.fix(word) for word in x.split()])
df.head(10)
```

```python
df['nospace'] = [' '.join(map(str,i)) for i in df['no_contract']]
df.head(10)
```

```python
df = df.drop('no_contract',axis=1)
```

```python
df.head(50)
```

```python
#tokenizing the tweets
df['tokenized'] = df['nospace'].apply(word_tokenize)
df.head(50)
```

```python
#removing punctuations
punc = string.punctuation
df['no_punct'] = df['tokenized'].apply(lambda x:[word for word in x if word not in punc])
df.head(50)
```

```python
#removing stopwords
stop_words = set(stopwords.words('english'))
df['sw_removed'] = df['no_punct'].apply(lambda x:[word for word in x if word not in stop_words])
df.head(10)
```

```python
#parts of speech tagging
df['pos_tags'] = df['sw_removed'].apply(nltk.tag.pos_tag)
df.head()
```

```python
def get_wordnet(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
    
df['wn_pos'] = df['pos_tags'].apply(lambda x:[(word,get_wordnet(pos_tag)) for (word,pos_tag) in x])
df.head(10)
```

```python
#lemmatizing words obtained using tokenization
wnl = WordNetLemmatizer()
df['lemmatized'] = df['wn_pos'].apply(lambda x : [wnl.lemmatize(word, tag) for word,tag in x])
df.head(10)
```

```python
df['string_lemma'] = [' '.join(map(str,i)) for i in df['lemmatized']]
```

```python
df.head(10)
```

```python
#exporting to a csv file for further lexicon based sentiment analysis
df.to_csv('lemmatized_data.csv')
```

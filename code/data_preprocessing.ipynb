# Data Preprocessing

*Removing noise from data*

```python
import pandas as pd
import re,string
import preprocessor as p
```

```python
tweets = pd.read_csv('text.csv',encoding = 'ISO-8859-1')
```

```python
pd.set_option('display.max_colwidth',-1)
tweets.head(10)
```

```python
tweets.info()
```

```python
#removing duplicates
tweets['is_duplicated'] = tweets.duplicated(['Text'])
tweets['is_duplicated'].sum()
```

```python
tweets.drop_duplicates(subset='Text',keep='first',inplace=True)
```

```python
tweets.head(10)
```

```python
tweets.drop('is_duplicated',axis=1)
```

```python
#to remove noise from data
def remove_noise(text):
    text = re.sub(r'<[^>]+>',"",text) # HTML tags
    text = re.sub(r'(?:@[\w_]+)',"",text) # @-mentions
    text = re.sub(r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)"," ",text) # hash-tags
    text = re.sub(r'(http[s]|www)|twitter?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+'," ",text,flags=re.MULTILINE) # URLs
    text = re.sub(r"(?:[\d]+)", "", text) # numbers
    text = re.sub('[!#?,.:";_]', ' ', text)# punctuation
    return text
tweets['cleaned_data'] = tweets.Text.apply(remove_noise)
```
